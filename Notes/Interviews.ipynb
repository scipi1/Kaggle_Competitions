```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
from sklearn.linear_model import LinearRegression

r = pickle.load(open("returns.pkl", "rb"))
r_train = r[:500]
r_test = r[500:-1]

#build feature and target vectors
X_train = np.zeros((480,20))
X_test = np.zeros((480,20))
Y_train = np.zeros((480,1))
Y_test = np.zeros((480,1))

for i,element in enumerate(r_train[20:-1]):
    Y_train[i] = element
    X_train[i,:] = r_train[i:i+20]
    
for i,element in enumerate(r_test[20:-1]):
    Y_test[i] = element
    X_test[i,:] = r_train[i:i+20]
    
reg = LinearRegression().fit(X_train, Y_train)
reg.score(X_train, Y_train)
reg.coef_
Y_result = reg.predict(X_test)
plt.plot(Y_test)
plt.plot(Y_result)
```
